```python
import pandas as pd;
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException


# 크롬 옵션 설정 (headless 모드 활성화)
chrome_options = Options()
# chrome_options.add_argument("--headless")  # 브라우저 창을 표시하지 않음


# Selenium 웹 드라이버 초기화
service = Service('./chromedriver-win64/chromedriver.exe')  # Chrome 드라이버 경로 설정
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
category_list = []
category_sub_list = []
author_href_list = []
img_author_src_list = []
name_author_list = []
label_category_list = []
title_post_list = []
text_list = []
```

```python
for page in range(1, 4):
    # 웹 페이지 열기
    url = "https://section.blog.naver.com/ThemePost.naver?directoryNo=29&activeDirectorySeq=3&currentPage=" + str(page)
    driver.get(url)
    print(driver.current_url)
    
    driver.implicitly_wait(2)

    # 주제별 요소 클릭
    posts = driver.find_elements(By.CSS_SELECTOR, '.info_post')

    for post in posts:
        # 카테고리
        category_list.append("취미·여가·여행")

        # 세부 주제
        category_sub_list.append("맛집")

        # 블로그 주소
        author = post.find_element(By.CSS_SELECTOR, '.author')
        author_href  = author.get_attribute('href')
        print(author_href)
        author_href_list.append(author_href)

        # 블로거 섬네일
        img_author = post.find_element(By.CSS_SELECTOR, '.img_author')
        img_author_src = img_author.get_attribute('src')
        print(img_author_src)
        img_author_src_list.append(img_author_src)

        # 블로거 정보
        name_author = post.find_element(By.CSS_SELECTOR, '.name_author')
        print(name_author.text)
        name_author_list.append(name_author.text)

        # 제목
        title_post = post.find_element(By.CSS_SELECTOR, '.title_post')
        print(title_post.text)
        title_post_list.append(title_post.text)

        # 내용
        text = post.find_element(By.CSS_SELECTOR, '.text')
        print(text.text)
        text_list.append(text.text)

        print(" ")

# 웹 브라우저 종료
# driver.quit()
```

```python
data = {"category": category_list, "category_sub": category_sub_list, "title_post": title_post_list, "text": text_list, "name_author": name_author_list, "author_href": author_href_list, "img_author_src": img_author_src_list}
df = pd.DataFrame(data)
print(df.tail(5))
```

```python
df.to_csv("category3.csv", encoding = "utf-8-sig")
```
